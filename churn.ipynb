```python
pf['TotalCharges'] = pd.to_numeric(pf['TotalCharges'], errors='coerce')
print("\nMissing after conversion:", pf['TotalCharges'].isnull().sum())

```

```python
pf = pf.dropna(subset=['TotalCharges'])
```

```python
print("\nSummary statistics:\n", pf.describe())

print("\nChurn Distribution:\n", pf['Churn'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=pf, x='Churn', palette='Set2')
plt.title("Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=pf, x='Contract', hue='Churn', palette='Set1')
plt.title("Contract vs Churn")
plt.show()
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data=pf, x=col, bins=30, kde=True, hue='Churn')
    plt.title(f"Distribution of {col} by Churn")
    plt.show()
```

Tenure means customer duration with the company

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=pf, x="Churn", y="MonthlyCharges", palette="coolwarm")
plt.title("Monthly Charges by Churn")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(pf[num_cols].corr(), annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

correlation heatmap is acolor coded table.which numerical features are positively or negatively related
